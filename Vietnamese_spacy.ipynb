```
import spacy
nlp = spacy.load('vi_spacy_model')
```

```
text = "Bộ luật Lao động quy định tiêu chuẩn lao động; quyền, nghĩa vụ, trách nhiệm của người lao động, người sử dụng lao động, tổ chức đại diện người lao động tại cơ sở, tổ chức đại diện người sử dụng lao động trong quan hệ lao động và các quan hệ khác liên quan trực tiếp đến quan hệ lao động; quản lý nhà nước về lao động."
doc = nlp(text)
```

```
import pandas as pd
```

```
tokens = []
for token in doc:
    tokens.append(
        {
         'token': token.text,
         'lemma': token.lemma_,
         'pos': token.pos_,
         'tag': token.tag_,
         'dep': token.dep_,
         'shape': token.shape_,
         'is_alpha': token.is_alpha,
         'is_stop': token.is_stop
        }
    )

df = pd.DataFrame(tokens)
df
```

```
options = {"fine_grained": True}
spacy.displacy.render(doc, style="dep", jupyter=True, options=options)
```
